Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Hardcoding some training values

```python
X = np.array([[1],[2],[3],[4]], dtype=float)
y = np.array([[2],[4],[6],[8]], dtype=float)
```

Add bias term to each sample

```python
X_b = np.c_[np.ones((X.shape[0],1)),X]
```

Initialize parameters

```python
theta = np.random.randn(2,1)
learning_rate = 0.01
n_iterations = 1000
m = len(X_b)
```

Gradient Descent

```python
cost_history = []
```

```python
for iteration in range(n_iterations):
    predictions = X_b.dot(theta)
    error = predictions - y
    gradients = (2/m) * X_b.T.dot(error)
    theta -= learning_rate * gradients
    cost = (1/m) * np.sum(error ** 2)
    cost_history.append(cost)
```

Output Learned Parameters

```python
print("Learned parameters (theta):" )
print(f"Intercept : {theta[0][0]:.4f}")
print(f"Slope : {theta[1][0]:.4f}")
print(theta)
```

Plot Loss Curve

```python
plt.plot(range(n_iterations), cost_history)
plt.title("Loss Curve")
plt.xlabel("Iterations")
plt.ylabel("MSE Cost")
plt.grid(True)
plt.show()
```

Plot Prediction Line

```python
X_new = np.array([[0], [6]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta)
```

```python
plt.scatter(X, y, color='blue', label='Training Data')
plt.plot(X_new, y_predict, color='red', label='Regression Line')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()
```
